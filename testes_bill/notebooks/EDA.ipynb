### 1. Configuração do Ambiente

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2. Visualização Geral do Dataset

```python
dataset = pd.read_csv('Telco-Customer-Churn.csv')
display(dataset)
```

<h4 style="font-weight:bold;" >Conteúdo</h4>
Cada linha representa um cliente, cada coluna contém atributos do cliente descritos nos metadados da coluna.

<h4 style="font-weight:bold;">O conjunto de dados inclui informações sobre:</h4>

* Clientes que saíram no último mês: a coluna é chamada de Churn (Evasão).
* Serviços que cada cliente contratou: telefone, múltiplas linhas, internet, segurança online, backup online, proteção de dispositivo, suporte técnico e streaming de TV e filmes.
* Informações da conta do cliente: há quanto tempo é cliente, contrato, método de pagamento, fatura digital (sem papel), cobranças mensais e cobranças totais.
* Informações demográficas sobre os clientes: gênero, faixa etária e se possuem parceiros e dependentes.

<h2 style="font-weight:bold;">Dicionário de Atributos:</h2>

* customerID: ID do cliente.
* gender: Gênero do cliente (Masculino/Feminino).
* SeniorCitizen: Indica se o cliente é idoso (1 para sim, 0 para não).
* Partner: Indica se o cliente possui parceiro ou parceira (Yes/No).
* Dependents: Indica se o cliente possui dependentes (Yes/No).
* tenure: Número de meses que o cliente permaneceu na empresa.
* PhoneService: Indica se o cliente possui serviço telefônico (Yes/No).
* MultipleLines: Indica se o cliente possui múltiplas linhas (Yes, No, No phone service).
* InternetService: Provedor de serviço de internet do cliente (DSL, Fiber optic, No).
* OnlineSecurity: Indica se o cliente tem segurança online (Yes, No, No internet service).
* OnlineBackup: Indica se o cliente tem backup online (Yes, No, No internet service).
* DeviceProtection: Indica se o cliente tem proteção de dispositivo (Yes, No, No internet service).
* TechSupport: Indica se o cliente tem suporte técnico (Yes, No, No internet service).
* StreamingTV: Indica se o cliente utiliza streaming de TV (Yes, No, No internet service).
* StreamingMovies: Indica se o cliente utiliza streaming de filmes (Yes, No, No internet service).
* Contract: O tipo de contrato do cliente (Month-to-month, One year, Two year).
* PaperlessBilling: Indica se o cliente utiliza fatura digital/online (Yes/No).
* PaymentMethod: O método de pagamento do cliente (Electronic check, Mailed check, Bank transfer, Credit card).
* MonthlyCharges: O valor cobrado do cliente mensalmente.
* TotalCharges: O valor total cobrado do cliente.
* Churn: Indica se o cliente cancelou o serviço e deixou a empresa (Yes/No).

### Análise de valores ausentes

```python
print("ANÁLISE DE MISSING VALUES\n")

missing_values = pd.DataFrame({
    'Coluna': dataset.columns,
    'Missing_Count': dataset.isnull().sum(),
    'Missing_Percentage': (dataset.isnull().sum() / len(dataset) * 100).round(2)
})

missing_values = missing_values[missing_values['Missing_Count'] > 0].sort_values(
    by='Missing_Percentage', ascending=False
)

if len(missing_values) > 0:
    print(missing_values)
    
    # Visualizar missing values
    plt.figure(figsize=(10, 6))
    plt.barh(missing_values['Coluna'], missing_values['Missing_Percentage'], color='coral')
    plt.xlabel('Porcentagem de Missing Values (%)')
    plt.title('Distribuição de Missing Values por Coluna')
    plt.tight_layout()
    plt.show()
else:
    print("Nenhum missing value detectado!")
```

### Verificar linhas duplicadas no dataset

```python
print("ANÁLISE DE DUPLICATAS\n")                                                                                                                                                                      
# Contar duplicatas considerando todas as colunas                                                                                                                                                     
duplicatas = dataset.duplicated()
numero_duplicatas = duplicatas.sum()

print(f"Número de linhas duplicadas: {numero_duplicatas}")
print(f"Percentual de duplicatas: {(numero_duplicatas / len(dataset) * 100):.2f}%\n")

# Se houver duplicatas, exibir algumas delas
if numero_duplicatas > 0:
    print("Primeiras 5 linhas duplicadas:")
    print(dataset[duplicatas].head())

    # Mostrar quantas vezes cada duplicata aparece
    print("\nContagem de duplicatas (as que aparecem mais de uma vez):")
    duplicatas_contagem = dataset[duplicatas].value_counts().head(10)
    print(duplicatas_contagem)
else:
    print("Nenhuma duplicata encontrada no dataset.")
```

## ANÁLISE DE PADRÕES, DISTRIBUIÇÕES, CORRELAÇÕES E TENDÊNCIAS

### 1. Análise da variável alvo (Churn)

```python
# Configurar estilo dos gráficos
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

print("1. DISTRIBUIÇÃO DA VARIÁVEL ALVO (Churn)")
churn_counts = dataset['Churn'].value_counts()
churn_percent = dataset['Churn'].value_counts(normalize=True) * 100

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
churn_counts.plot(kind='bar', ax=ax1, color=['skyblue', 'lightcoral'])
ax1.set_title('Contagem de Churn')
ax1.set_xlabel('Churn')
ax1.set_ylabel('Número de Clientes')

churn_percent.plot(kind='bar', ax=ax2, color=['skyblue', 'lightcoral'])
ax2.set_title('Percentual de Churn')
ax2.set_xlabel('Churn')
ax2.set_ylabel('Percentual (%)')

for i, v in enumerate(churn_percent):
      ax2.text(i, v + 1, f'{v:.1f}%', ha='center')

plt.tight_layout()
plt.show()

print(f"Distribuição do Churn:\n{churn_percent}\n")
```

### 2. Análise de distribuições de variáveis numéricas

```python
print("2. DISTRIBUIÇÕES DE VARIÁVEIS NUMÉRICAS")

num_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, col in enumerate(num_cols):
    # Converter TotalCharges para numérico se necessário
    if col == 'TotalCharges':
        data = pd.to_numeric(dataset[col], errors='coerce')
    else:
        data = dataset[col]

    # Plotar o histograma com a curva de densidade (KDE)
    sns.histplot(data=data, kde=True, ax=axes[i])
    axes[i].set_title(f'Distribuição de {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequência')

plt.tight_layout()
plt.show()
```

### 3. Análise de matriz de correlações. 

```python
print("3. MATRIZ DE CORRELAÇÕES\n")

# Etapa 1: Preparar dados numéricos (sem alterar o dataset original)
dataset_numeric = dataset.copy()

# Converter TotalCharges para numérico (tratando valores não numéricos como NaN)
dataset_numeric['TotalCharges'] = pd.to_numeric(dataset_numeric['TotalCharges'], errors='coerce')

# Garantir que SeniorCitizen seja numérico
dataset_numeric['SeniorCitizen'] = dataset_numeric['SeniorCitizen'].astype(float)

# Selecionar APENAS as colunas numéricas que queremos correlacionar
numeric_cols = ['SeniorCitizen', 'tenure', 'MonthlyCharges', 'TotalCharges']
dataset_numeric = dataset_numeric[numeric_cols].copy()

# Etapa 2: Calcular e visualizar a matriz de correlação
correlation_matrix = dataset_numeric.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(
    correlation_matrix, 
    annot=True, 
    cmap='coolwarm', 
    center=0,
    square=True, 
    linewidths=0.5, 
    cbar_kws={"shrink": .8}
)

plt.title('Matriz de Correlação das Variáveis Numéricas')
plt.tight_layout()
plt.show()

# Etapa 3: Correlações específicas com o Churn (variável alvo)
# Criar uma cópia apenas para este cálculo, trazendo o Churn como numérico
dataset_churn = dataset.copy()
dataset_churn['TotalCharges'] = pd.to_numeric(dataset_churn['TotalCharges'], errors='coerce')
dataset_churn['Churn_num'] = dataset_churn['Churn'].map({'Yes': 1, 'No': 0})

# Calcular correlações com Churn_num
churn_correlations = dataset_churn[numeric_cols + ['Churn_num']].corr()['Churn_num'].sort_values(key=abs, ascending=False)

print("Top 10 correlações com Churn (variável alvo):")
print(churn_correlations.head(11))  # Inclui o próprio Churn_num
print()
```

### 4. Análise por categorias importantes

```python
print("4. ANÁLISE DE CHURN POR CATEGORIAS IMPORTANTES")

cat_cols = ['Contract', 'InternetService', 'PaymentMethod', 'SeniorCitizen', 'Partner', 'Dependents']

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for i, col in enumerate(cat_cols):
    if col in dataset.columns:
        # Criar tabela de contingência
        churn_by_cat = pd.crosstab(dataset[col], dataset['Churn'], normalize='index') * 100

        # Plotar
        churn_by_cat.plot(kind='bar', stacked=True, ax=axes[i])
        axes[i].set_title(f'Taxa de Churn por {col}')
        axes[i].set_xlabel(col)
        axes[i].set_ylabel('Percentual (%)')
        axes[i].legend(title='Churn', loc='upper right')
        axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

### 5. Análise de tendência ao longo do tempo (tenure)

```python
print("5. ANÁLISE DE TENDÊNCIA AO LONGO DO TEMPO (Tenure)")

# Usar uma cópia para não afetar o dataset original
dataset_tenure = dataset.copy()
dataset_tenure['TotalCharges'] = pd.to_numeric(dataset_tenure['TotalCharges'], errors='coerce')

# Criando grupos de tenure (faixas etárias de contrato)
bins = [0, 12, 24, 36, 48, 60, 72]
labels = ['0-12', '13-24', '25-36', '37-48', '49-60', '61-72']

dataset['tenure_group'] = pd.cut(dataset['tenure'], bins=bins, labels=labels)

# Calculando a taxa de churn por período de tenure
churn_by_tenure = pd.crosstab(
    dataset['tenure_group'], 
    dataset['Churn'], 
    normalize='index'
) * 100

# Configuração e exibição do gráfico
churn_by_tenure.plot(kind='bar', stacked=True, color=['skyblue', 'lightcoral'])

plt.title('Taxa de Churn por Grupo de Tenure')
plt.xlabel('Tenure (meses)')
plt.ylabel('Percentual (%)')
plt.legend(title='Churn', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### 6. Análise de MonthlyCharges por churn  (possivel outlier)

```python
print("6. COMPARAÇÃO DE MonthlyCharges ENTRE CHURN E NÃO-CHURN")

# Configuração do gráfico de Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Churn', y='MonthlyCharges', data=dataset)

plt.title('Distribuição de MonthlyCharges por Status de Churn')
plt.xlabel('Churn')
plt.ylabel('MonthlyCharges')
plt.tight_layout()
plt.show()

# Exibição das estatísticas descritivas
print("Estatísticas de MonthlyCharges por Churn:")
print(dataset.groupby('Churn')['MonthlyCharges'].describe())
print()
```

 Com base na análise exploratória, podemos formular as seguintes hipóteses de negócio:

* 1: Clientes com contrato mês a mês têm significativamente maior taxa de churn do que aqueles com contratos de longo prazo.
* 2: Clientes com serviço de fibra óptica têm maior probabilidade de churn do que aqueles com DSL.
* 3: Clientes que utilizam pagamento por cheque eletrônico têm maior taxa de churn.
* 4: Houve um aumento no churn entre clientes com tenure entre 12-24 meses (possível período de reavaliação de contrato).
* 5: Clientes com altos MonthlyCharges baixos tendem a churn mais (sensibilidade ao preço).
* 6: Clientes sem parceiro ou dependentes têm maior probabilidade de churn.
